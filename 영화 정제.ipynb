```python
df5 = df4.copy()

def process_rottentomatoes(df):
    # 1. 로튼토마토 수치 컬럼 생성
    df['로튼토마토_수치'] = df['로튼토마토'].str.replace('%', '')
    df['로튼토마토_수치'] = pd.to_numeric(df['로튼토마토_수치'], errors='coerce')
    
    # 2. 로튼토마토 평가여부 컬럼 생성
    df['로튼토마토_평가여부'] = df['로튼토마토'].apply(lambda x: 0 if x == '- %' else 1)
    
    return df

process_rottentomatoes(df5)
```

```python
df8 = df7.copy()

```

```python
# 장르 전처리 및 원-핫 인코딩 함수
def process_genres(df):
    # 1. 문자열을 리스트로 변환
    df['장르'] = df['장르'].apply(lambda x: eval(x) if isinstance(x, str) else x)
    
    # 2. 유사/중복 장르 통합
    genre_mapping = {
        '멜로/로맨스': '로맨스',
        '공포(호러)': '공포',
        '어드벤처(모험)': '어드벤처',
        '서부극(웨스턴)': '웨스턴'
    }
    
    df['장르'] = df['장르'].apply(lambda x: [genre_mapping.get(g, g) for g in x] if isinstance(x, list) else x)
    
    # 3. 원-핫 인코딩
    # 모든 고유 장르 추출
    all_genres = set()
    for genres in df['장르']:
        if isinstance(genres, list):
            all_genres.update(genres)
    
    # 각 장르에 대한 열 생성
    for genre in sorted(all_genres):
        df[f'장르_{genre}'] = df['장르'].apply(lambda x: 1 if isinstance(x, list) and genre in x else 0)
    
    # 4. 기존 장르 열 삭제
    df = df.drop('장르', axis=1)
    
    return df

# 함수 적용
df8 = process_genres(df8)
```

유사/중복 장르를 통합하여 데이터를 단순화
각 장르를 개별 열로 만들어 분석이 용이

## 개봉일 열 정제 

```python
df10 = df9.copy()
```

```python
df10.info()
```

```python
df10[df10['개봉일'].isna()] # 누락정보 확인
```

```python
df10 = df10.fillna({'개봉일' : '없음'}) # 개봉일의 누락값을 없음으로 바꾸기
```

```python
df10['개봉일'].value_counts()
```

```python
def split_release_dates(date_str):
    if '재개봉' in date_str:
        # 재개봉일과 최초 개봉일 분리
        rerelease_date, original_date = date_str.split(' 재개봉 ')
        rerelease_date = rerelease_date.replace(' 개봉', '')
        original_date = original_date.replace(' 개봉', '')
        return pd.Series([original_date, rerelease_date])
    else:
        # 일반 개봉일인 경우
        return pd.Series([date_str.replace(' 개봉', ''), None])

# 개봉일과 재개봉일로 분리
df10[['개봉일', '재개봉일']] = df10['개봉일'].apply(split_release_dates)

# datetime 형식으로 변환
for col in ['개봉일', '재개봉일']:
    df10[col] = pd.to_datetime(df10[col], format='%Y년 %m월 %d일', errors='coerce')
    df10[col] = df10[col].dt.strftime('%Y-%m-%d')

# '없음' 값 처리
df10.loc[df10['개봉일'].isna(), '개봉일'] = '없음'
```


이 코드는 다음과 같이 데이터를 처리합니다:

1. '개봉일'과 '재개봉일' 두 개의 열로 분리:
   - 예시 입력: "2024년 02월 08일 재개봉 2004년 02월 20일 개봉"
   - '개봉일': "2004-02-20" (최초 개봉일)
   - '재개봉일': "2024-02-08" (재개봉일)

2. 일반 개봉작의 경우:
   - '개봉일': 해당 개봉일자
   - '재개봉일': None (또는 NaN)

이렇게 처리하면:
- 최초 개봉일과 재개봉일 정보를 모두 보존할 수 있습니다
- 개봉일과 재개봉일을 각각 분석할 수 있습니다
- 재개봉 여부를 쉽게 파악할 수 있습니다 (재개봉일 컬럼이 NULL이 아닌 경우)

데이터 확인을 위해 다음 코드를 실행해볼 수 있습니다:

```python
# 재개봉 영화만 확인
rerelease_movies = df10[df10['재개봉일'].notna()]
print("재개봉 영화 수:", len(rerelease_movies))
print("\n재개봉 영화 목록:")
print(rerelease_movies[['작품명', '개봉일', '재개봉일']])
```
